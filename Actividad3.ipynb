**Conceptos basicos de la libreria MATPLOTLIB**
Matplotlib es una libreria de trazado de graficos de bajo nivel en python que sirve como una herramienta de visualizacion. Matplotlib es de codigo abierto y podemos utilizarlo libremente. Matplotlib esta escrito principalmente en python con algunos segmentos escritos en C
-----------
**Instalacion de matplotlib**
Si tienes python y pip ya instalado en tu sistema, instala matplotlib usando el siguiente comando   **pip install matplotlib**
-----------
**Importar matplotlib**
Una vez instalado matplotlib importalo a tus aplicaciones agregando el modulo import, ejemplo: **import matplotlib***
-----------
**Pyplot**
la mayoria de las herramientas de matplotlib se encuentran dentro del submodulo pyplot y normalmente se importa bajo el alias plt
**import matplotlib.pyplot as plt**
Ahora el paquete pyplot puede denominarse plt
